```python
import pandas as pd
```

```python
df_1 = pd.read_csv("/content/despesa_ceaps_2022.csv", skiprows=[0], sep=";")
df_2 = pd.read_csv("/content/despesa_ceaps_2021.csv", skiprows=[0], sep=";")
df_3 = pd.read_csv("/content/despesa_ceaps_2020.csv", skiprows=[0], sep=";")
df_4 = pd.read_csv("/content/despesa_ceaps_2019.csv", skiprows=[0], sep=";")
```

```python
df_anos = pd.concat([df_1, df_2, df_3, df_4])
df_anos
```

```python
df_anos.info()
```

```python
# Convert 'DATA' column to datetime objects first
df_anos["DATA"] = pd.to_datetime(df_anos["DATA"], format="%d/%m/%Y", errors='coerce')

# Or to convert to Unix timestamp (seconds since epoch):
df_anos["Timestamp"] = df_anos["DATA"].astype("int64") // 10**9
```

```python
df_anos.info()
```

```python
df_anos
```

```python
df_anos = df_anos.drop(columns=["COD_DOCUMENTO", "DOCUMENTO", "Timestamp"])
```

```python
df_anos.isna().sum()
```

```python
df_anos[df_anos.isnull().any(axis=1)]
```

```python
df_anos = df_anos.fillna("não detalhado")
```

```python
df_anos.query("DETALHAMENTO == 'não detalhado'")
```

```python
df_anos
```

```python
df_anos["VALOR_REEMBOLSADO"] = pd.to_numeric(df_anos["VALOR_REEMBOLSADO"], errors="coerce")
despesas_senador = df_anos.groupby(["SENADOR"])["VALOR_REEMBOLSADO"].sum().sort_values(ascending=False).head(10)
despesas_senador = despesas_senador.reset_index()
despesas_senador
```

```python
despesas_senador.info()
```

```python
import matplotlib.pyplot as plt
```

```python
cores=[]

fig, ax = plt.subplots(figsize=(10,6))
ax.barh(despesas_senador["SENADOR"], despesas_senador["VALOR_REEMBOLSADO"], color="silver")
ax.set_title("Despesas dos 10 principais senadores (milhões)", loc="left", fontsize=18)
ax.set_xlabel("Valor das despesas")
ax.set_ylabel("")

for i, v in enumerate(despesas_senador["VALOR_REEMBOLSADO"]):
  ax.text(v+1000, i, str(v), color="Black", fontsize=10, ha="left", va="center" )

#excluir caixa do grafico (linhas externar)
ax.set_frame_on(False)

#excluir linha vertical (x)
ax.get_xaxis().set_visible(False)

#removendo os marcadores
ax.tick_params(axis="both", which="both", length=0)

plt.show()


```
